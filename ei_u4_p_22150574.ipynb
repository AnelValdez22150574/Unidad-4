#***INTRODUCCIN***

El rendimiento acadmico de los estudiantes en los exmenes es un indicador crucial para evaluar la efectividad de los sistemas educativos y el potencial futuro de los jvenes. Sin embargo, este desempeo no est determinado nicamente por habilidades individuales, sino que depende de una variedad de factores interrelacionados que influyen en el aprendizaje. Entre estos factores se incluyen los hbitos de estudio, la asistencia regular a clases, la participacin activa de los padres en el proceso educativo, el acceso a recursos de aprendizaje, las horas dedicadas al estudio y al sueo, as como las calificaciones previas. Adems, elementos como el nivel de motivacin, el acceso a internet, las sesiones de tutora, el ingreso familiar, la calidad del profesorado, y el tipo de escuela (pblica o privada) desempean roles significativos en el rendimiento acadmico.

El contexto social y las actividades extracurriculares tambin tienen un impacto, ya que influyen en el desarrollo integral del estudiante. Por otro lado, factores fsicos como la actividad fsica y las discapacidades de aprendizaje, junto con caractersticas personales como el gnero, la distancia entre el hogar y la escuela, y la influencia de compaeros, son variables que no deben ser ignoradas. Asimismo, el nivel educativo de los padres, el ingreso familiar y la calidad del profesorado contribuyen a configurar el entorno de aprendizaje del estudiante.

Este estudio estadstico busca analizar cmo estos factores afectan el rendimiento de los estudiantes en los exmenes, tomando como punto de partida el puntaje obtenido y correlacionndolo con las variables mencionadas. Al identificar patrones y relaciones significativas entre estos factores, se espera proporcionar informacin valiosa para la toma de decisiones en polticas educativas y estrategias pedaggicas, con el objetivo de optimizar el rendimiento acadmico y garantizar una educacin ms equitativa y de calidad.

#***MARCO TERICO***

El rendimiento acadmico de los estudiantes es un tema ampliamente estudiado en la literatura educativa, dado su impacto en el desarrollo personal y social, as como en la formacin del capital humano. Este marco terico explora los principales factores que influyen en el desempeo de los estudiantes en los exmenes, agrupndolos en categoras: individuales, familiares, escolares y sociales.

**1. Factores Individuales**
Los factores individuales incluyen caractersticas personales y hbitos que afectan directamente el aprendizaje. Segn Zimmerman (2002), los hbitos de estudio, como la planificacin y la organizacin del tiempo, son predictores significativos del rendimiento acadmico. Del mismo modo, las horas de sueo juegan un papel esencial en la memoria y la atencin (Walker, 2009). El nivel de motivacin del estudiante, entendido como el esfuerzo sostenido hacia metas especficas, tambin influye considerablemente en los resultados acadmicos (Deci & Ryan, 1985). Adems, el historial de calificaciones previas puede reflejar tanto habilidades adquiridas como patrones de aprendizaje consolidados.

**2. Factores Familiares**
La participacin de los padres en el proceso educativo es un factor determinante. Epstein (2001) sostiene que el involucramiento parental positivo fomenta un ambiente de aprendizaje en el hogar y genera expectativas acadmicas altas. Asimismo, el nivel educativo de los padres y el ingreso familiar afectan la disponibilidad de recursos y oportunidades de aprendizaje, ya que influyen en el acceso a tutoras, materiales didcticos y tecnologa (Coleman, 1966).

**3. Factores Escolares**
La calidad del profesorado y el tipo de escuela (pblica o privada) son variables fundamentales. Estudios como el de Hanushek (1997) indican que los profesores con experiencia y formacin adecuada generan un impacto directo en el rendimiento de los estudiantes. Adems, el acceso a recursos pedaggicos y la infraestructura escolar influyen en la experiencia educativa. Las sesiones de tutora tambin han mostrado ser efectivas en la mejora del desempeo, al proporcionar apoyo personalizado (Bloom, 1984).

**4. Factores Sociales y Contextuales**
El entorno social del estudiante, incluyendo la influencia de sus compaeros y su participacin en actividades extracurriculares, tambin afecta el rendimiento. Bandura (1977) destaca la importancia del aprendizaje social, donde los estudiantes modelan comportamientos y actitudes observadas en sus pares. Por otro lado, la actividad fsica ha sido relacionada con mejoras en la cognicin y el manejo del estrs (Hillman et al., 2008). Factores como la distancia entre el hogar y la escuela y la conectividad a internet tambin determinan el tiempo y la calidad del aprendizaje.

**5. Inclusin y Diversidad**
La presencia de discapacidades de aprendizaje plantea desafos adicionales en el rendimiento acadmico. Segn el modelo social de la discapacidad (Oliver, 1990), es crucial adaptar el entorno educativo para satisfacer las necesidades de los estudiantes y garantizar su inclusin plena.

**Marco Conceptual**

Rendimiento acadmico: desempeo de los estudiantes medido a travs de puntajes obtenidos en evaluaciones.
Hbitos de estudio: patrones y estrategias adoptados para organizar y optimizar el aprendizaje.
Participacin parental: grado de involucramiento de los padres en la educacin de sus hijos.
Calidad del profesorado: nivel de preparacin y eficacia de los docentes.
Factores contextuales: condiciones externas al individuo, como recursos disponibles y entorno social.
Este marco terico proporciona una visin integral que permite entender cmo interactan estos factores en el rendimiento acadmico de los estudiantes, sirviendo como base para el anlisis y la interpretacin de los resultados del estudio estadstico.

#***METODOLOGA***

El presente estudio estadstico se orienta a analizar el impacto de mltiples factores en el rendimiento acadmico de los estudiantes en exmenes mediante la aplicacin del anlisis de varianza (ANOVA) y la prueba de comparaciones mltiples de Tukey.

**1. Diseo del Estudio**

Se emplear un diseo cuantitativo no experimental, de tipo correlacional-comparativo. Se busca identificar diferencias significativas entre las medias de los puntajes de exmenes en funcin de diferentes niveles de los factores categricos considerados.


*   *Variable dependiente:* puntaje obtenido en los exmenes.
*   *Variables independientes:* factores categricos como hbitos de estudio, nivel de motivacin, participacin parental, calidad del profesorado, tipo de escuela, gnero, actividades extracurriculares, nivel educativo de los padres, entre otros.

**2. Anlisis estadstico**


*   ***Preparacin de datos:***

*Validacin y limpieza de datos:* se eliminarn valores atpicos extremos y datos incompletos.

*Verificacin de supuestos para ANOVA:*

Normalidad de la distribucin (prueba de Shapiro-Wilk).

Homogeneidad de varianzas.

Independencia de las observaciones.

*   ***Aplicacin del ANOVA***

El nivel de significancia se establecer en  = 0.05.

Hiptesis nula (H): No existen diferencias significativas en el rendimiento acadmico entre los niveles de los factores independientes.
Hiptesis alternativa (H): Existen diferencias significativas en el rendimiento acadmico entre los niveles de los factores independientes.

*   ***Prueba de Comparaciones Mltiples (Tukey)***

En caso de rechazar la hiptesis nula, se aplicar la prueba de Tukey para determinar cules grupos presentan diferencias significativas. Esta prueba comparar las medias de los grupos categricos ajustando el nivel de error para las mltiples comparaciones.

**Interpretacin de Resultados**

Los resultados del ANOVA se interpretarn en funcin del valor p, mientras que las comparaciones de Tukey se analizarn a travs de los intervalos de confianza y las diferencias significativas entre medias.

```python
# Si todos los grupos tienen una distribucin normal
import scipy.stats as stats

nivel_de_significancia = 0.05

# Realizar ANOVA
anova_result = stats.f_oneway(
    df[df['horas_estudio'] == '19>']['puntaje_examen'],
    df[df['horas_estudio'] == '<20,>34']['puntaje_examen'],
    df[df['horas_estudio'] == '<35']['puntaje_examen'],

)

# Interpretar los resultados
print("Resultados del ANOVA:")
print(f"F-Estadstico: {anova_result.statistic:.2f}")
print(f"Valor p: {anova_result.pvalue:.5f}\n")

if anova_result.pvalue < nivel_de_significancia:
    print("Conclusin: Hay diferencias significativas entre al menos dos tratamientos.")
else:
    print("Conclusin: No hay diferencias significativas entre los tratamientos.")
```

```python
from statsmodels.stats.multicomp import pairwise_tukeyhsd
import matplotlib.pyplot as plt

nivel_de_significancia = 0.05

# Prueba de Tukey
tukey = pairwise_tukeyhsd(endog=df['puntaje_examen'], groups=df['horas_estudio'], alpha=nivel_de_significancia)

# Mostrar los resultados
print(tukey)

# Grfico de las diferencias entre grupos
tukey.plot_simultaneous(ylabel="Grupos", xlabel="Medias")

plt.gca().spines['right'].set_visible(False) # derecha
plt.gca().spines['top'].set_visible(False)   # superior
plt.title("Mltiples comparaciones entre todos los pares (Tukey)")
```
