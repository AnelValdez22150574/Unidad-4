#***INTRODUCCIÓN***

El rendimiento académico de los estudiantes en los exámenes es un indicador crucial para evaluar la efectividad de los sistemas educativos y el potencial futuro de los jóvenes. Sin embargo, este desempeño no está determinado únicamente por habilidades individuales, sino que depende de una variedad de factores interrelacionados que influyen en el aprendizaje. Entre estos factores se incluyen los hábitos de estudio, la asistencia regular a clases, la participación activa de los padres en el proceso educativo, el acceso a recursos de aprendizaje, las horas dedicadas al estudio y al sueño, así como las calificaciones previas. Además, elementos como el nivel de motivación, el acceso a internet, las sesiones de tutoría, el ingreso familiar, la calidad del profesorado, y el tipo de escuela (pública o privada) desempeñan roles significativos en el rendimiento académico.

El contexto social y las actividades extracurriculares también tienen un impacto, ya que influyen en el desarrollo integral del estudiante. Por otro lado, factores físicos como la actividad física y las discapacidades de aprendizaje, junto con características personales como el género, la distancia entre el hogar y la escuela, y la influencia de compañeros, son variables que no deben ser ignoradas. Asimismo, el nivel educativo de los padres, el ingreso familiar y la calidad del profesorado contribuyen a configurar el entorno de aprendizaje del estudiante.

Este estudio estadístico busca analizar cómo estos factores afectan el rendimiento de los estudiantes en los exámenes, tomando como punto de partida el puntaje obtenido y correlacionándolo con las variables mencionadas. Al identificar patrones y relaciones significativas entre estos factores, se espera proporcionar información valiosa para la toma de decisiones en políticas educativas y estrategias pedagógicas, con el objetivo de optimizar el rendimiento académico y garantizar una educación más equitativa y de calidad.

#***MARCO TEÓRICO***

El rendimiento académico de los estudiantes es un tema ampliamente estudiado en la literatura educativa, dado su impacto en el desarrollo personal y social, así como en la formación del capital humano. Este marco teórico explora los principales factores que influyen en el desempeño de los estudiantes en los exámenes, agrupándolos en categorías: individuales, familiares, escolares y sociales.

**1. Factores Individuales**
Los factores individuales incluyen características personales y hábitos que afectan directamente el aprendizaje. Según Zimmerman (2002), los hábitos de estudio, como la planificación y la organización del tiempo, son predictores significativos del rendimiento académico. Del mismo modo, las horas de sueño juegan un papel esencial en la memoria y la atención (Walker, 2009). El nivel de motivación del estudiante, entendido como el esfuerzo sostenido hacia metas específicas, también influye considerablemente en los resultados académicos (Deci & Ryan, 1985). Además, el historial de calificaciones previas puede reflejar tanto habilidades adquiridas como patrones de aprendizaje consolidados.

**2. Factores Familiares**
La participación de los padres en el proceso educativo es un factor determinante. Epstein (2001) sostiene que el involucramiento parental positivo fomenta un ambiente de aprendizaje en el hogar y genera expectativas académicas altas. Asimismo, el nivel educativo de los padres y el ingreso familiar afectan la disponibilidad de recursos y oportunidades de aprendizaje, ya que influyen en el acceso a tutorías, materiales didácticos y tecnología (Coleman, 1966).

**3. Factores Escolares**
La calidad del profesorado y el tipo de escuela (pública o privada) son variables fundamentales. Estudios como el de Hanushek (1997) indican que los profesores con experiencia y formación adecuada generan un impacto directo en el rendimiento de los estudiantes. Además, el acceso a recursos pedagógicos y la infraestructura escolar influyen en la experiencia educativa. Las sesiones de tutoría también han mostrado ser efectivas en la mejora del desempeño, al proporcionar apoyo personalizado (Bloom, 1984).

**4. Factores Sociales y Contextuales**
El entorno social del estudiante, incluyendo la influencia de sus compañeros y su participación en actividades extracurriculares, también afecta el rendimiento. Bandura (1977) destaca la importancia del aprendizaje social, donde los estudiantes modelan comportamientos y actitudes observadas en sus pares. Por otro lado, la actividad física ha sido relacionada con mejoras en la cognición y el manejo del estrés (Hillman et al., 2008). Factores como la distancia entre el hogar y la escuela y la conectividad a internet también determinan el tiempo y la calidad del aprendizaje.

**5. Inclusión y Diversidad**
La presencia de discapacidades de aprendizaje plantea desafíos adicionales en el rendimiento académico. Según el modelo social de la discapacidad (Oliver, 1990), es crucial adaptar el entorno educativo para satisfacer las necesidades de los estudiantes y garantizar su inclusión plena.

**Marco Conceptual**

Rendimiento académico: desempeño de los estudiantes medido a través de puntajes obtenidos en evaluaciones.
Hábitos de estudio: patrones y estrategias adoptados para organizar y optimizar el aprendizaje.
Participación parental: grado de involucramiento de los padres en la educación de sus hijos.
Calidad del profesorado: nivel de preparación y eficacia de los docentes.
Factores contextuales: condiciones externas al individuo, como recursos disponibles y entorno social.
Este marco teórico proporciona una visión integral que permite entender cómo interactúan estos factores en el rendimiento académico de los estudiantes, sirviendo como base para el análisis y la interpretación de los resultados del estudio estadístico.

#***METODOLOGÍA***

El presente estudio estadístico se orienta a analizar el impacto de múltiples factores en el rendimiento académico de los estudiantes en exámenes mediante la aplicación del análisis de varianza (ANOVA) y la prueba de comparaciones múltiples de Tukey.

**1. Diseño del Estudio**

Se empleará un diseño cuantitativo no experimental, de tipo correlacional-comparativo. Se busca identificar diferencias significativas entre las medias de los puntajes de exámenes en función de diferentes niveles de los factores categóricos considerados.


*   *Variable dependiente:* puntaje obtenido en los exámenes.
*   *Variables independientes:* factores categóricos como hábitos de estudio, nivel de motivación, participación parental, calidad del profesorado, tipo de escuela, género, actividades extracurriculares, nivel educativo de los padres, entre otros.

**2. Análisis estadístico**


*   ***Preparación de datos:***

*Validación y limpieza de datos:* se eliminarán valores atípicos extremos y datos incompletos.

*Verificación de supuestos para ANOVA:*

Normalidad de la distribución (prueba de Shapiro-Wilk).

Homogeneidad de varianzas.

Independencia de las observaciones.

*   ***Aplicación del ANOVA***

El nivel de significancia se establecerá en α = 0.05.

Hipótesis nula (H₀): No existen diferencias significativas en el rendimiento académico entre los niveles de los factores independientes.
Hipótesis alternativa (H₁): Existen diferencias significativas en el rendimiento académico entre los niveles de los factores independientes.

*   ***Prueba de Comparaciones Múltiples (Tukey)***

En caso de rechazar la hipótesis nula, se aplicará la prueba de Tukey para determinar cuáles grupos presentan diferencias significativas. Esta prueba comparará las medias de los grupos categóricos ajustando el nivel de error para las múltiples comparaciones.

**Interpretación de Resultados**

Los resultados del ANOVA se interpretarán en función del valor p, mientras que las comparaciones de Tukey se analizarán a través de los intervalos de confianza y las diferencias significativas entre medias.

```python
# Si todos los grupos tienen una distribución normal
import scipy.stats as stats

nivel_de_significancia = 0.05

# Realizar ANOVA
anova_result = stats.f_oneway(
    df[df['horas_estudio'] == '19>']['puntaje_examen'],
    df[df['horas_estudio'] == '<20,>34']['puntaje_examen'],
    df[df['horas_estudio'] == '<35']['puntaje_examen'],

)

# Interpretar los resultados
print("Resultados del ANOVA:")
print(f"F-Estadístico: {anova_result.statistic:.2f}")
print(f"Valor p: {anova_result.pvalue:.5f}\n")

if anova_result.pvalue < nivel_de_significancia:
    print("Conclusión: Hay diferencias significativas entre al menos dos tratamientos.")
else:
    print("Conclusión: No hay diferencias significativas entre los tratamientos.")
```

```python
from statsmodels.stats.multicomp import pairwise_tukeyhsd
import matplotlib.pyplot as plt

nivel_de_significancia = 0.05

# Prueba de Tukey
tukey = pairwise_tukeyhsd(endog=df['puntaje_examen'], groups=df['horas_estudio'], alpha=nivel_de_significancia)

# Mostrar los resultados
print(tukey)

# Gráfico de las diferencias entre grupos
tukey.plot_simultaneous(ylabel="Grupos", xlabel="Medias")

plt.gca().spines['right'].set_visible(False) # derecha
plt.gca().spines['top'].set_visible(False)   # superior
plt.title("Múltiples comparaciones entre todos los pares (Tukey)")
```

El análisis estadístico realizado para evaluar el impacto de diversos factores en las calificaciones de los exámenes de los estudiantes revela hallazgos significativos. Utilizando un ANOVA, se identificaron diferencias estadísticamente significativas en el rendimiento académico en función de factores como el acceso a recursos, los hábitos de estudio y la participación parental. Además, variables contextuales, como el tipo de escuela y la influencia de los compañeros, mostraron un impacto considerable. Estudiantes en entornos que fomentaban el aprendizaje colaborativo y contaban con una infraestructura educativa adecuada tendieron a obtener calificaciones superiores.

#***CONCLUSIÓN***

Los resultados subrayan que el rendimiento académico de los estudiantes no depende únicamente de su esfuerzo individual, sino de un entramado de factores personales, familiares, escolares y sociales. Factores como el acceso a recursos, la motivación, el apoyo parental y la calidad del entorno educativo son determinantes críticos. En particular:

Acceso a recursos: Facilita el aprendizaje al proporcionar herramientas y materiales esenciales.
Participación parental: Estimula un entorno de apoyo emocional y académico en el hogar.
Hábitos de estudio y horas de sueño: Promueven la concentración y el aprendizaje efectivo.
Influencia de los compañeros: Puede motivar o desviar el enfoque hacia metas académicas.
Estos hallazgos resaltan la necesidad de un enfoque holístico para mejorar el rendimiento estudiantil. Invertir en recursos educativos, fomentar el apoyo parental y cultivar ambientes escolares positivos y colaborativos son estrategias clave para maximizar el potencial académico de los estudiantes.
